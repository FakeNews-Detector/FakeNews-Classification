# Web Scrapping

```python
# Requirements Installation
!pip install beautifulsoup4
!pip install requests

# Import libraries
import bs4
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Getting HTML Content and save it to coverpage
rl = requests.get("https://turnbackhoax.id/category/fitnah-hasut-hoax/")
coverpage = rl.content
print(coverpage)
```

```python
soup1 = BeautifulSoup(coverpage, "html.parser")
```

```python
coverpage_news = soup1.find_all("h3", attrs = {"class" : "entry-title mh-loop-title"})
print(coverpage_news)
```

```python
print(len(coverpage_news))
```

```python
frame = []
upperframe = []
for j in coverpage_news:
  title_news = j.find("a").get_text().strip()
  link_news = j.find("a")["href"].strip()
  frame.append([title_news, link_news])
upperframe.extend(frame)
```

```python
data = pd.DataFrame(upperframe, columns=["title_news","link_news"])
data
```

```python
# Scrapping the page from 1-100
URL = "https://turnbackhoax.id/category/fitnah-hasut-hoax/page/"
data = []
upperframe = []

for page in range(1,100):
  req = requests.get(URL + str(page) + "/")
  soup = BeautifulSoup(req.text, "html.parser")

  coverpage_news = soup.find_all("h3", attrs = {"class" : "entry-title mh-loop-title"})
  for j in coverpage_news:
    judul = j.find("a").get_text().strip()
    link = j.find("a")["href"].strip()
    data.append([judul, link])
  upperframe.extend(data)  
```

```python
data = pd.DataFrame(upperframe, columns=["judul","link"])
data.head(10)
```

```python
data["kategori"] = "Hoax"
data 
```

```python
data.to_csv("Data_News_Hoax.csv")
```
